create table Schedule (
	grade char(32) references Grades(name) not null,
	room char(32) references Rooms(name) not null,
	teacher char(32) references Teachers(initials) not null,
	term int references Terms(year) not null,
	day DAYS not null,
	time tsrange not null,
	foreign key (day, time) references Lessons (day, time),
	UNIQUE(day, time, teacher, term), /* one teacher teaches only one lesson at a time */
	UNIQUE(day, time, room, term), /* one room hosts only one lesson at a time */
	UNIQUE(day, time, grade, term) /* one grade can attend only one lesson at a time */
);

```python
df_rooms = df_csv[['Days','Lesson_Times','AKC','DMS','BL','BS']][1::2]
df_rooms = df_rooms.fillna(0)
```

```python
df_grades = df_csv[['Days','Lesson_Times','AKC','DMS','BL','BS']][::2]
df_grades = df_grades.fillna(0)
```

```python
#ROOMS FOR TEACHERS

AKC_rooms = df_rooms.AKC.values.tolist()
DMS_rooms = df_rooms.DMS.values.tolist()
BL_rooms = df_rooms.BL.values.tolist()
BS_rooms = df_rooms.BS.values.tolist()

AKC_rooms = [i for i in AKC_rooms if i != 0]
DMS_rooms = [i for i in DMS_rooms if i != 0]
BL_rooms = [i for i in BL_rooms if i != 0]
BS_rooms = [i for i in BS_rooms if i != 0]
```

```python
#GRADES FOR TEACHERS

AKC_grade = df_grades.AKC.values.tolist()
DMS_grade = df_grades.DMS.values.tolist()
BL_grade = df_grades.BL.values.tolist()
BS_grade = df_grades.BS.values.tolist()

AKC_grade = [i for i in AKC_grade if i != 0]
DMS_grade = [i for i in DMS_grade if i != 0]
BL_grade = [i for i in BL_grade if i != 0]
BS_grade = [i for i in BS_grade if i != 0]
```

```python
AKC_room = df_rooms[['Days','Lesson_Times', 'AKC']]
AKC_room = AKC_room[AKC_room['AKC'] != 0]
#AKC_room.insert(0, 'Name', 'AKC')

DMS_room = df_rooms[['Days','Lesson_Times', 'DMS']]
DMS_room = DMS_room[DMS_room['DMS'] != 0]
#DMS_room.insert(0, 'Name', 'DMS')

BL_room = df_rooms[['Days','Lesson_Times', 'BL']]
BL_room = BL_room[BL_room['BL'] != 0]
#BL_room.insert(0, 'Name', 'BL')

BS_room = df_rooms[['Days','Lesson_Times', 'BS']]
BS_room = BS_room[BS_room['BS'] != 0]
#BS_room = BS_room.insert(0, 'Name', 'BS')
```

```python
AKC_grade = df_grades[['Days','Lesson_Times', 'AKC']]
AKC_grade = AKC_grade[AKC_grade['AKC'] != 0]


DMS_grade = df_grades[['Days','Lesson_Times', 'DMS']]
DMS_grade = DMS_grade[DMS_grade['DMS'] != 0]

BL_grade = df_grades[['Days','Lesson_Times', 'BL']]
BL_grade = BL_grade[BL_grade['BL'] != 0]

BS_grade = df_grades[['Days','Lesson_Times', 'BS']]
BS_grade = BS_grade[BS_grade['BS'] != 0]

#AKC_grade.insert(0, 'Name', 'AKC')
#DMS_grade.insert(0, 'Name', 'DMS')
#BL_grade.insert(0, 'Name', 'BL')
#BS_grade.insert(0, 'Name', 'BS')
```

```python
#SCHEDULE FOR EACH TEACHER

AKC = AKC_room.merge(AKC_grade, on='Days')
AKC.rename(columns={'Days': "Days", 'Lesson_Times_x': "Lesson_Times", 'AKC_x': "Room", 'Lesson_Times_y':"Lesson_Number",'AKC_y':"Grade"}, inplace=True)

DMS = DMS_room.merge(DMS_grade, on='Days')
DMS.rename(columns={'Days': "Days", 'Lesson_Times_x': "Lesson_Times", 'DMS_x': "Room", 'Lesson_Times_y':"Lesson_Number",'DMS_y':"Grade"}, inplace=True)

BL = BL_room.merge(BL_grade, on='Days')
BL.rename(columns={'Days': "Days", 'Lesson_Times_x': "Lesson_Times", 'BL_x': "Room", 'Lesson_Times_y':"Lesson_Number",'BL_y':"Grade"}, inplace=True)

BS = BS_room.merge(BS_grade, on='Days')
BS.rename(columns={'Days': "Days", 'Lesson_Times_x': "Lesson_Times", 'BS_x': "Room", 'Lesson_Times_y':"Lesson_Number",'BS_y':"Grade"}, inplace=True)
```

```python
create table Students (
	id integer primary key,
	name char(64),
	attend char(32) references Grades(name) 
		not null	/* total participation */
);
```

```python
def insert_Students():
    ids = []
    names = []
    grades = []
    for i in range(5, 11):
        for j in range(ord('A'), ord('E')):
            grades.append(str(i) + chr(j))
    for i in range(1, 3):
        ids.append(int(i))
    for i in range(ord('A'), ord('Z')):
        for j in range(ord('A'), ord('Z')):
            names.append(chr(i) + chr(j))
    for i in ids:
        for grade in grades:
            for name in names: 
                print("insert into Students values ('{}','{}','{}');".format(i,name,grade))
```
